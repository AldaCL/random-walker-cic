```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def bound_truc(k):
    if 999 - k < 0:
        k = 999
    elif 999 - k > 999:
        k = 0
    else:
        None
    return k
```

```python
grid = np.zeros([1000, 1000])
```

```python
# 1. Inicialmente con dos números aleatorios se obtine la posición inicial
x_0 = random.randint(0, 999)
y_0 = random.randint(0, 999)

grid[x_0, y_0] = 1
```

```python
# 2. Dirección, 8 posibilidades
x = x_0
y = y_0

# contador de pixeles visitados
area = 0

n = 1000000 # número de iteraciones

for _ in range(n):
    # 3. Distancia a moverse (aleatoria)
    directions = [
    (0, 1),    # Arriba
    (1, 1),    # Arriba derecha
    (1, 0),    # Derecha
    (1, -1),   # Abajo derecha
    (0, -1),   # Abajo
    (-1, -1),  # Abajo izquierda
    (-1, 0),   # Izquierda
    (-1, 1)    # Arriba izquierda
    ]
    
    direction = random.randint(0, 999)
    mov = 1000 / 8

    mov_indx = int(direction // (mov))
    dx, dy = directions[mov_indx]

    p = random.randint(0, 10) # tamaño del paso [0, 10]

    x += dx * p
    y += dy * p

    # Si está fuera de las fronteras se trunca al valor de la pared
    x = bound_truc(x)
    y = bound_truc(y)

    # 4. Marca el área visitada
    grid[x, y] += 1

    if grid[x, y] == 1:
        area += 1

    # 5. Porcentaje de área visitada
    prob = area / (1000 * 1000)
    
print(f'Porcentaje de área visitada: {prob * 100} %')
```

```python
plt.imshow(grid, cmap='jet')
plt.show()
```
